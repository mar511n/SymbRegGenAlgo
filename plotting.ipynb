```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
complexityMeasures = np.load("complexityMeasures.npy")
print(complexityMeasures.shape)
```

```python
c_edges = np.linspace(0, 1, 11)
c_hist = complexityMeasures[:]
bin_centers = (c_edges[:-1] + c_edges[1:]) / 2
fig, ax = plt.subplots(figsize=(7, 4))
ax.imshow(c_hist[:,::-1].T, extent=[0, len(c_hist), np.min(c_edges), np.max(c_edges)], aspect='auto', cmap='viridis')
ax.set_xlabel('Time (Generations)')
ax.set_ylabel('Complexity Measure')
plt.show()
```
